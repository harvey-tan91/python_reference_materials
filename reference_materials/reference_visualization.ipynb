***Matplotlib Reference***
==========================
```import matplotlib.pyplot as plt```

*Reference Materials*
---------------------
* [API - matplotlib](https://matplotlib.org/stable/api/index.html)
* [API - plt.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
    * [API - plt.plot function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot)
<br>

*Basic*
-------
* plt.style.available --- to return all the available matplotlib style
* plt.style.use('...)
<br>
* plt.plot(x, y
            color= 'r', 
            linestyle= '-',
            marker= 'o',
            drawstyle= 'steps-post',
            label= z) --- to create a line chart
* plt.bar() --- vertical bar chart
* plt.barh() --- horizontal bar chart
* plt.scatter()

*Chart Information*
-------------------
* plt.figure((10,10)) --- to adjust the size of the chart figure
* plt.legend(loc=df.X1) --- display legend
* plt.xlabel( rotation={0,1}, fontsize=..., y=... ) --- to display the x-label title
* plt.ylabel() --- to display the y-label title
* plt.grid(True) --- to display the grid
* plt.title('...') --- to display the chart title
* plt.axis([x,x,y,y]) --- to set the min and max value of the x and y coordinates
    * plt.xlim( (int, int) )
    * plt.xlim( (int, int) )
* plt.text(x,y, 'text') --- to insert a text box at the defined x,y coordinates
* plt.savefig('name of file.png') --- this will save a copy of the chart in the cwd

*Creating Figures with Subplots*
--------------------------------
* fig, ( (ax1, ax2), (ax3, ax4) ) = plt.subplots(nrows=2, ncols=2, figsize=...)
    * ax1.plot(...)
    * ax1.scatter(...)
        * [Reference](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes)   
        * ax1.set_title('this is the title')
        * ax1.set_xlim(0,10000)
        * ax1.set_xscale('linear) --- set the type of scaling
            * value: {'linear', 'log', 'symlog', 'logit', ...} og
        * ax.set_xlabel('this is label') --- set title for x-axis
        * ax1.invert_xaxis()

*Plotting from Pandas DataFrame*
--------------------------------
* pd.plotting.autocorrelation_plot(data.X1)
* df.hist(fig_size=...) --- return the histogram of all numerical attributes
* df.X1.plot()

*Updating or Modifying Figures made with Plotly Express*
--------------------------------------------------------
Plotly Express create Figure objects.

**Figures Method**
------------------

**Update Layout**
```
fig.update_layout(title_text='...', title_font_color='red', title_font_size=30,
                    font_family='Rockwell', 
                    height=1000, width=1000)
```
**Update X-axis**
```
fig.update_xaxes(title='kkk', 
                    type= {'linear', 'log'}, showgrid= {True, False}
                    zeroline= {True, False})
```
**Update Y-axis**
```
fig.update_yaxes()
```
**Adding New Shape**
```
fig.add_shape()
```
**Adding Vertical Line**
```
fig.addvline(x=1, line_width=3, line_color, line_dash='dash'
                    annotation_text = '...', annotation_position = 'top right')
```
**Adding Horizontal Line**
```
fig.add_hline(y=1,..........)
```
**Adding Vertical Rectangle**
```
fig.add_vrect(x0=0.5, x1=2.5, line_width=0, fillcolor='red', opacity)
```
**Adding Horizontal Rectangle**
```
fig.add_hrect(y0, y1, ...)
```

*How to create Figure Object*
-----------------------------

```
go.Figure() ---> figure object

plotly.subplots.make_subplots(rows, cols, shared_xaxes={False, True, 'rows', 'all'}, shared_yaxes,
                                horizontal_spacing= [0,1], vertical_spacing = [0,1],
                                subplot_titles = (...), 
                                x_title = 'text at the bottom of the chart', y_title,
                                row_titles, column_titles
                                ) ---> figure object
```
